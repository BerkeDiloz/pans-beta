```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

countries = ['denmark', 'netherlands', 'germany', 'sweden',
             'poland', 'belgium', 'slovakia', 'hungary',
             'austria', 'spain', 'czech']

counts = [100, 100, 90, 80, 70, 80, 10, 60, 100, 90, 90]

bar_labels = ['red', 'orange', 'green', 'blue',
              'red', 'orange', 'green', 'blue',
              'red', 'orange', 'green']

bar_colors = ['tab:red', 'tab:orange', 'tab:green', 'tab:blue',
              'tab:red', 'tab:orange', 'tab:green', 'tab:blue',
              'tab:red', 'tab:orange', 'tab:green']

ax.bar(countries, counts, color=bar_colors)

ax.set_ylabel('visited countries')
ax.set_title('visited countries')

plt.show()
```
