```python
import matplotlib.pyplot as plt

# 1. Prepare categorical data
categories = ['MPL', 'BM', 'MR', 'SM', 'M', 'BM2','MMR']
quantities = [75, 55, 40, 20, 50, 35, 25]

# 2. Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(categories, quantities, color=['red', 'yellow', 'darkred', 'green', 'pink', 'black', 'darkgreen', 'darkblue'])

# 3. Customize the chart
plt.title('Difficulty Rates of Lessons')
plt.xlabel('Lesson Types')
plt.ylabel('')

# Add exact values on top of the bars
for i, value in enumerate(quantities):
    plt.text(i, value+0.5, str(value), ha='center')

plt.show()
```
