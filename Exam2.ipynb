```python
# 1. Creating my random list of numbers
numbers = [25, 42, 8, 91, 73, 15, 66, 4, 88, 50, 33, 12]

# 2. Calculations

average = sum(numbers) / len(numbers)
biggest = max(numbers)
smallest = min(numbers)

# 3. Printing results
print(f"Numbers List: {numbers}")
print(f"Average: {average:.2f}")
print(f"Biggest value is: {biggest}")
print(f"Smallest value is: {smallest}")
```
