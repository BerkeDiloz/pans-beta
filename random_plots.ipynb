```python
import random
import matplotlib.pyplot as plt


random_values = [random.randint(1, 10) for _ in range(12)]
months = list(range(1, 13))


plt.plot(months, random_values, marker='x', linestyle='-', color='r')


plt.xlabel('Months')
plt.ylabel('Values')
plt.title('The Random Values')
plt.xticks(months)
plt.grid(True)


plt.show()
```

```python
import random
import matplotlib.pyplot as plt


random_values = [random.randint(1, 10) for _ in range(12)]
months = list(range(1, 13))


plt.barh(months, random_values, color='b', edgecolor='black')


plt.xlabel('Values')
plt.ylabel('Months')
plt.title('The Random Values')


plt.yticks(months)


plt.grid(True, axis='x', linestyle='--', alpha=0.7)


plt.show()
```
