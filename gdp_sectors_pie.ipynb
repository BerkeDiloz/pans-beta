```python
import matplotlib.pyplot as plt

# 1. GRAFİK: GSYİH (GDP) Karşılaştırması
ulkeler = ['Türkiye', 'İspanya']
gdp_degerleri = [1118, 1620]

plt.pie(gdp_degerleri, labels=ulkeler, autopct='%1.1f%%')
plt.title('GSYİH Karşılaştırması')
plt.show()  # İlk grafiği ekranda gösterir

# 2. GRAFİK: Türkiye Ekonomik Sektörleri
sektorler = ['Sanayi', 'Tarım', 'Hizmet']
sektor_degerleri = [30, 7, 63]

plt.pie(sektor_degerleri, labels=sektorler, autopct='%1.1f%%')
plt.title('Türkiye Ekonomik Sektörleri')
plt.show()  # İkinci grafiği ekranda gösterir
```
