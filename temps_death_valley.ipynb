```python
import pandas as pd

def far2cel(f):
    return int(round((f-32)*5/9,0))

df = pd.read_csv('https://raw.githubusercontent.com/andrzejmp/some_codes/refs/heads/main/python/temperatures/data.csv')

highs = df["TMAX"]
lows = df["TMIN"]

highs_c=[]
lows_c=[]
for x in highs:
    highs_c.append(far2cel(x))
for x in lows:
    lows_c.append(far2cel(x))


import matplotlib.pyplot as plt
plt.figure(figsize=(14, 8))

plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.plot(highs_c, c='red')
ax.plot(lows_c, c='blue')
ax.set_title('Max and min temps in The Death Valley, 2018', fontsize=16)
ax.set_ylabel('Temp. in C', fontsize=12)
ax.set_xlabel('Days', fontsize=12)
#print(plt.style.available)
plt.show()
```
