```python
import matplotlib.pyplot as plt
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "doctors" ,
    "programmers",
    "teachers" ,
)
weight_counts = {
    "Below": np.array([160, 55, 60]),
    "Above": np.array([150, 145, 30]),
}
width = 0.8

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("the structure of incomes of these groups in thousands TRY")
ax.legend(loc="upper right")

plt.show()
```
