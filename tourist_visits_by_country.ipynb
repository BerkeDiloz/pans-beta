```python
import matplotlib.pyplot as plt

# 1. Kategoriler aylardan ülkelere değiştirildi
categories = ['Türkiye', 'Fransa', 'İspanya', 'İtalya']
numbers = []

# Veriyi dosyadan okuma
with open('data.csv', 'r', encoding='utf-8') as file:
    content = file.read()
    nums = content.split(',')
    for n in nums:
        # .strip() z
        numbers.append(int(n.strip()))

# 2. Bar grafiğini oluşturma
plt.figure(figsize=(10, 6))

# Kodunuzun sonundaki renk listesini bar grafiğine ekledim
colors = ['red', 'yellow', 'darkred', 'brown']
plt.bar(categories, numbers, color=colors)

# 3. Grafiği özelleştirme
plt.title('Ülkelere Göre Ziyaret Eden Turist Sayısı')
plt.xlabel('Ülkeler')
plt.ylabel('Turist Sayısı (Milyon / Kişi)')


for i, value in enumerate(numbers):
    plt.text(i, value + (max(numbers)*0.01), str(value), ha='center')

plt.show()
```
