```python
import matplotlib.pyplot as plt
import random

data = [random.randint(1, 100) for _ in range(30)]

#plt.figure(figsize=(10, 6))
plt.plot(data, marker='*', linestyle=':', color='r')
plt.title('Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

data = [random.randint(1, 100) for _ in range(30)]

plt.bar(range(1, 31), data, color='red', edgecolor='black')
plt.title('Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(axis='x', linestyle='--', alpha=0.9)

plt.show()
```

```python
import matplotlib.pyplot as plt
import random

data = [random.randint(1, 100) for _ in range(30)]


plt.barh(range(1, 31), data, color='red', edgecolor='black')
plt.title('Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(axis='y', linestyle=':', alpha=0.9)
```
