```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-4, 4, 100)
y1 = 2-x*x

# Tworzenie wykresu
fig, ax = plt.subplots(figsize=(8, 5))

# Ukrycie osi Y
ax.spines['left'].set_visible(True)
ax.get_yaxis().set_visible(True)

# Oś X jako linia z końcową strzałką
ax.axhline(0, color='black', linewidth=1)
ax.arrow(10, 0, 0.5, 0, head_width=0.08, head_length=0.3, fc='black', ec='black', clip_on=False)

# Wykresy funkcji
ax.plot(x, y1, color='blue', linestyle='-', linewidth=2)
#ax.plot(x, y2, color='black', linestyle='-', linewidth=2)


# Etykieta osi X
ax.text(10.7, -0.1, "x", fontsize=12)

# Ustawienia siatki, tytułu i granic
ax.set_title("Wykresy funkcji y = arctg(x) (przerywana) i y = -arctg(x) (ciągła)")
ax.grid(True, linestyle='--', linewidth=0.5, axis='x')
ax.set_xlim(-4, 3)
ax.set_ylim(-3, 3)

plt.show()
```
