```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare the data
x = np.linspace(0, 10, 100) # 100 evenly spaced points between 0 and 10
y1 = np.sin(x)
y2 = np.cos(x)

# 2. Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='Sine Wave', color='blue', linewidth=2)
plt.plot(x, y2, label='Cosine Wave', color='red', linestyle='--')

# 3. Add labels and title
plt.title('Line Plot Example: Sine and Cosine Waves')
plt.xlabel('X-axis (Time)')
plt.ylabel('Y-axis (Amplitude)')

# 4. Add a legend and grid
plt.legend()
plt.grid(True, alpha=0.3)

# 5. Display the plot
plt.show()
```
