```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("D:\\Python\\Articles\\matplotlib\\sales_data.csv")

labels = [
    "FaceCream",
    "FaceWash",
    "ToothPaste",
    "Bathing soap",
    "Shampoo",
    "Moisturizer",
]
salesData = [
    df["facecream"].sum(),
    df["facewash"].sum(),
    df["toothpaste"].sum(),
    df["bathingsoap"].sum(),
    df["shampoo"].sum(),
    df["moisturizer"].sum(),
]

plt.figure(figsize=(8, 6))
plt.pie(salesData, labels=labels, autopct="%1.1f%%", startangle=140)
plt.axis("equal")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.title("Sales data")
plt.tight_layout()
plt.show()
```
